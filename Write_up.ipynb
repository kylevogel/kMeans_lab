Well Mr. Rooney, after working through the lab, I finished the report you asked for on best valued NBA players you should look into. These are players who statistically perform better than their peers for the salaries that they have, and would provide substantially more value to your team for a cheaper amount. Before I tell you exactly who those players are, I want to walk you through the main steps of how I came to this conclusion.

### Step 1 - Organizing and Preparing the Data

I started out with two datasets that were incomplete and had errors that I needed to fix before I could trust anything. This included extra weird characters around players’ names, rows where the player name wasn’t even there, and the biggest issue which was players showing up multiple times because they were traded and played for multiple teams. To solve that, I cleaned the player name column so it would match correctly across the two datasets, removed the rows where the player name was missing, and then for traded players I only kept the stats from the team they played the most games for. That way, every player only showed up once, and it represented their main role for the season.

On the salary side, the file loaded in a weird format where the header information was mixed into the data and the columns were labeled as “Unnamed.” I fixed the column names and dropped the first row so that the salary data was clean. Then I converted salary from a string like “$70,732” into an actual numeric dollar amount.

After both datasets were cleaned, I merged them into one dataset using the player name as the join key, so every player had both stats and salary in the same row.

### Step 2 - Choosing the Stats that Best Define Perfomance

Next, I needed to decide what performance metrics I wanted to use to group players. I narrowed it down to total points, total rebounds, total assists, total steals, total blocks, total turnovers, games started, and minutes played. I chose these because they give a simple but pretty complete snapshot of who is producing, who is defending, who is handling the ball, and who is actually getting real minutes and opportunity.

### Step 3 - Grouping Players via Clustering

![Cluster results: Scoring vs Rebounding, Scoring vs Turnovers, Scoring vs Minutes Played](cluster_scatterplots.png)

Once I chose the stats, I scaled them so they would be similarly important to each other. The reason scaling matters is because the raw ranges of these stats are totally different, and without scaling, something like minutes played would dominate the model just because it has bigger numbers.

Then I ran KMeans clustering to group players into similar statistical profiles. The whole point of this step is that it doesn’t make sense to compare a low minute bench player to a high minute starter. Clustering lets me compare players to the right type of peers first, and then judge salary inside that peer group.

### Step 4 - Determining the Number of Clusters

![Silhouette and inertia results for different K values](silhouette_k_test.png)

![Elbow method plot for choosing K](elbow_plot.png)

After running KMeans, I evaluated the clustering using two checks. I used silhouette score to see how clean and separated the clusters were, and I also used inertia through the elbow method to see when adding more clusters stopped improving things by much. The elbow looked around 3, but 4 gave me slightly cleaner separation (with a higher silhouette) while still keeping the results easy to interpret, so I went with 4 clusters. It also gave me tighter clusters than 3 (lower inertia) and slightly better separation (higher silhouette). One thing that is a point of emphasis in clustering is that it is not saying one player is “good” or “bad.” It is just grouping players who look statistically similar. That matters because it lets me compare salary in a fair way. For example, it would not be fair to compare a bench player to a full time starter, since their roles and minutes are totally different. The clustering here helps create better comparisons.

### Step 5 - Finding the Best Valued Players

![Most underpriced players compared to their cluster average salary](Recommendations.png)

Once every player had a cluster label, I used the clusters to compare salary fairly. I calculated the average salary within each cluster, then compared each player’s salary to that average. If a player’s salary was way below the cluster average, that means they are cheaper than other players with similar production, and that is exactly the type of player you told me you want to target.

From there, I sorted players by how much cheaper they were than their cluster average and pulled the best value options. This method is not perfect, but it gives a clean starting point. A player can look underpriced in the data for reasons that have nothing to do with talent, like injury, a rookie contract, or a role change. So I would treat this as a shortlist of names to research further, not a final answer by itself.
